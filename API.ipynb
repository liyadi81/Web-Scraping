```python
import flickrapi
import numpy as np
import pandas as pd
import csv
# api_key = '***********************'
# api_secret = '**************'
flickr = flickrapi.FlickrAPI(api_key, api_secret)
photolist = pd.read_csv('photo2.csv')
photolist1 = photolist.url.str.replace('https://www.flickr.com/photos/', '')
photolist2 = photolist1.str.split('/', expand=True)
photoidlist = photolist2[1]
csv_file = open('photoinfo_API.csv', 'a+', newline='')
writer = csv.writer(csv_file)
writer.writerow(['owner_name', 'owner_location', 'photo_title', 'n_views', 'n_comments', 'date_taken', 'camera_brand',
                'camera_model', 'exposure', 'apeture', 'ISO', 'focal_length', 'flash', 'lens'])
i = 0
for photoid in photoidlist:
    photo_dict = {}    
    try:
        photo_exif = flickr.photos.getExif(photo_id = photoid, format = "parsed-json")
    except:
        print('no camera info')    
    try:
        photo_info = flickr.photos.getInfo(photo_id = photoid, format = "parsed-json")
    except:
        print('photo missing')
        continue    
    photo_dict['owner_name'] = photo_info['photo']['owner']['realname'].encode('utf-8')
    photo_dict['owner_location'] = photo_info['photo']['owner']['location'].encode('utf-8')
    photo_dict['photo_title'] = photo_info['photo']['title']['_content'].encode('utf-8')
    photo_dict['n_views'] = photo_info['photo']['views']
    photo_dict['n_comments'] = photo_info['photo']['comments']['_content']
    photo_dict['date_taken'] = photo_info['photo']['dates']['taken']       
    for exif in photo_exif['photo']['exif']:
        if exif.get('label') == 'Make':
            photo_dict['camera_brand'] = exif['raw']['_content']
        if exif.get('label') == 'Model':
            photo_dict['camera_model'] = exif['raw']['_content']
        if exif.get('label') == 'Exposure':
            photo_dict['exposure'] = exif['raw']['_content']
        if exif.get('label') == 'Aperture':
            photo_dict['aperture'] = exif['raw']['_content']
        if exif.get('label') == 'ISO Speed':
            photo_dict['ISO'] = exif['raw']['_content']
        if exif.get('label') == 'Focal Length':
            photo_dict['focal_length'] = exif['raw']['_content']
        if exif.get('label') == 'Flash':
            photo_dict['flash'] = exif['raw']['_content']
        if exif.get('label') == 'Lens Model':
            photo_dict['lens'] = exif['raw']['_content']    
    writer.writerow(photo_dict.values())
    print("scraped {} th photo, id: {}".format(i+1, photoid))
    i = i + 1
    if i > 1780:
        print("stopped after photoid: {}".format(photoid))
        csv_file.close()
        break
    
```
